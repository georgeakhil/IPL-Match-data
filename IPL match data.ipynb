### 2.   When should you use an entity versus attribute?

Ans. An entity is a real-world object that represents data in RDBMS. The entities in our model are Match, Team, Player, Venue  and Results. Attributes are properties that describe an entity. For example, the venue ID, venue name and city are the attributes of the entity Venue. In our model, match could have been an entity and an attribute. We decide to use it as an entity because it is relational to Team, Venue and Result. Using Match as an entity prevented data redundancy.

### 3.   When should you use an entity or relationship, and placement of attributes?

Ans. When there is a relationship between the attributes within an entity. In our model, the attributes Player_ID and Player_captain_ID are related within the entity Player.


### 4.   How did you choose your keys? Which are unique?

Ans. We have a total of 5 tables/entities and we have one primary in each table. For our first table, ‘Player’, we have chosen ‘Player_ID’ as our primary key as it is a unique number associated with each player, there would be no repetitions. In the ‘Team’ table, we have taken ‘Team_ID’ as our primary key as every team will have a single unique ID only. This is connected to the Foreign Key ‘Team_ID’ in the ‘Player’ table with a ‘many-to-one and only one’ relationship as there can be many team ID’s but a player can only be a part of one team. ‘Match_ID’ is the Primary Key for the ‘Match’ table. It is connected to the Primary Key from the ‘Result’ table which is a combination of two columns; ‘Match_ID’ and ‘Match_Winner’. Similarly, ‘Venue_ID’ is the Primary Key of the ‘Venue’ table.

### 5.   Did you model hierarchies using the “ISA” design element? Why or why not?

Ans. No. Since my model in this assignment is a bit simple and the attributes and entities are all unique and do not need hierarchal inheritance. So it’s unnecessary to use ISA design element.

### 6.   Were there design alternatives? What are their tradeoffs: entity vs. attribute, entity vs. relationship, binary vs. ternary relationships?

Ans. Yes, there were. We have used a composite primary key in the ‘Result’ table. Instead of this, we could’ve created another attribute called ‘Result_ID’ and used it as the primary key and used ‘Match_ID’ as the foreign key. Also, in the ‘Team’ table, we could’ve used the attribute ‘Team_Name’ itself as the primary key as it is unique, but we’ve assigned a Team_ID and used this as the primary key. This helps us make our tables occupy less space but forces us to add another attribute.
